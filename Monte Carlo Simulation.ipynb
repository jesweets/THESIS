```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib . pyplot as plt
from scipy . optimize import root , minimize
sns . set_style ('whitegrid')
```

```python
p = 1
n = 1
N=23
# generates the payment pattern
pay_lst = []
while n <= 23:
    pay = np . random . binomial (1 , p )
    pay_lst . append ( pay )
    if pay == 1:
        n += 1
    else :
        continue

def get_equation ( q ) :
    qsum = 0
    for i in range (len ( pay_lst ) ) :
        if pay_lst [ i ] == 1:
            qsum += (1.10) *( q **( i +1) )
        else :
            continue
    return (N - qsum )

sol = root ( get_equation, 1)
Q = float ( sol . x )
R = float ( - N * np . log ( sol . x ) )

num_microloans = 10000
p = .20 # probability of payment
R_lst = []
R_lst_0 = []
R_lst_1 = []
for i in range ( num_microloans ) :
    n = 1
    pay_lst = []
    while n <= 23:
        pay = np . random . binomial (1 , p )
        pay_lst . append ( pay )
        if pay == 1:
            n += 1
        else :
            continue
    R = float ( - N * np . log ( root ( get_equation ,1) . x ) )
    if len ( pay_lst ) == 23:
        R_lst_0 . append ( R )
    elif len ( pay_lst ) == 24:
        R_lst_1 . append ( R )
    R_lst . append ( R )
df = pd . DataFrame ( data ={" R_lst ": R_lst })
df

plt . hist ( R_lst , bins = 50)
plt . xlabel (" Interest Rate (R)")
plt . ylabel (" Number of Occurences ")
# plt. savefig ( 'p_25 .png', dpi =300 , bbox_inches = 'tight')
#plt. hist ( R_lst_4 , bins = 20)
bins = np . linspace (.15 , .19 , 23)
plt . legend ( loc ='upper left')
plt . xlabel (" Interest Rate (R)")
plt . ylabel (" Number of Occurences ")
#plt. savefig ( ' diffdelays .png ', dpi =300 , bbox_inches = ' tight ')

df . describe () . style . format ('{: ,} ')
```
